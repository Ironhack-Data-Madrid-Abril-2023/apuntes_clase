```python
def check_nan(df: pd.DataFrame) -> None:
    
    """Recibe un dataframe y enseña el % de valores nulos y lo grafica"""
    
    nan_cols = df.isna().mean() * 100  # porcentaje de nulo en cada columna
    
    display(f'N nan cols: {len(nan_cols[nan_cols>0])}')
    display(nan_cols[nan_cols>0])
    
    
    # grafico de nulos en el dataframe
    plt.figure(figsize=(10, 6))     # inicia la figura y establece el tamaño
    sns.heatmap(df.isna(),          # datos
                yticklabels=False,  # quita las etiquetas del eje y
                cmap='viridis',     # mapa de color
                cbar=False,         # sin barra lateral
               )


    plt.show();
```

```python
def limpiar(string):
    
    try:
        return string.split('-')  # lista con año, mes, dia
    
    except:
        return [np.nan, np.nan, np.nan]
```

## Valores Nulos

```python
check_nan(df)
```

```python
# umbral del 20%, columnas con mas del 20% de nulos fuera

nan_cols=df.isna().mean() * 100

nan_cols[nan_cols>20]
```

```python
df = df.drop(columns=nan_cols[nan_cols>20].index)

check_nan(df)
```

```python
# desglosado

num_nan = []    # nº de nulos por fila

for fila in df.itertuples():
    
    check = []
    
    for e in fila:
        
        check.append(pd.isna(e))
        
    num_nan.append(sum(check))
    
    
num_nan[:5]
```

```python
max(num_nan)
```

```python
pd.isna(np.nan)
```

```python
df['num_nan'] = [sum([pd.isna(e) for e in fila]) for fila in df.itertuples()]

df.head()
```

```python
check_nan(df[df.num_nan>=10])
```

```python
# umbral de 10 por fila, si hay mas de 10 nulos borro la fila

df[df.num_nan>=10].head()
```

```python
df = df[df.num_nan<10]

df.shape
```

```python
check_nan(df)
```

```python
df.boxplot();
```

```python
df.fillna(df.median(), inplace=True)
```
