```python
import matplotlib.pyplot as plt
import numpy as np


%matplotlib inline
```

```python
def square(L, c=(0, 0)):
    x=(np.array([0,0,1,1, 0])*L)-c[0]-L/2
    y=(np.array([0,1,1,0, 0])*L)-c[1]-L/2
    return np.array([x, y])
```

```python
def square_rot(L, center=(0,0), ro=0):
    c=square(L, center)
    
    rot=np.array([
        [np.cos(ro), -np.sin(ro)],
        [np.sin(ro), np.cos(ro)]
    ])
    
    res=[]
    for i in range(len(c[0, :])):
        q=np.dot(rot, c[:, i])
        res.append(q)
        
    return np.array(res).T
```

```python
# xkcd.com

with plt.xkcd():               # estilo xkcd
    plt.figure(figsize=(8,8))  # establecer figura y tamao
    
    index_ticks=list(range(-20, 20, 4))                           # indices/etiquetas eje x
    plt.xticks(index_ticks, index_ticks, rotation=35)             # cambio de xticks con rotacion
    
    for i in range(100):                            # 100 cuadrados
        c=square_rot(10-i, (0,0), (np.pi/50)*i)
        plt.plot(*c, color='#336699', alpha=1-(i/50))
    
plt.title('Cuadrados girando')      # titulo
plt.xlabel('Coordenada X')          # titulos de coordenadas
plt.ylabel('Coordenada Y')
plt.legend(['espiral'], loc='best') # leyenda, best es donde sea sin tapar grafico
plt.savefig('cuadrados.png')
plt.show();
```
